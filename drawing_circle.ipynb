```python
import matplotlib.pyplot as plt
def create_circle():
    circle=plt.Circle((0,0), radius =0.5, fc='g',ec='r')
    return circle
```

```python
def show_shape(patch):
    ax=plt.gca()
    ax.add_patch(patch)
    plt.axis('scaled')
    plt.show()
```

```python
#if __name__== '_main_':
c=create_circle()
show_shape(c)
```
