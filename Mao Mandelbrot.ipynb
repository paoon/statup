```python
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import random
```

```python
xmin = -2.5
xmax = 1.0
ymin = -1.0
ymax = 1.0
```

```python
def initialize_image(x_p, y_p):
    image=[]
    for i in range(y_p):
        x_colors=[]
        for j in range(x_p):
            x_colors.append(0)
        image.append(x_colors)
    return image
```

```python
def getCordinates(i, k):
    xi = xmin + (xmax - xmin) * i / 400
    yk = ymin + (ymax - ymin) * k / 400
    return xi, yk
```

```python
def getIter(x, y):
    max_iteration=100
    z=0+0j
    J = 1j
    c=x+y*J
    iteration=0
    while abs(z)<4 and iteration<max_iteration:
        z=z**2+c
        iteration=iteration+1
    return iteration
```

```python
def MandIter(i, k):
    xi, yk = getCordinates(i, k)
    iterNum = getIter(xi, yk)
    return iterNum
```

```python
def color_points():
    x_p=400
    y_p=400
    image=initialize_image(x_p, y_p)
    for i in range(y_p):
        for j in range(x_p):
             image[k][i] = MandIter(i, k)
    plt.imshow(image, origin='lower', extent=(-2.5, 1.0,-1.0, 1.0), cmap=cm.Greys_r, interpolation='nearest')
    plt.colorbar()
    plt.show()
```

```python
color_points()
```
