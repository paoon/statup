```python
import random
import matplotlib.pyplot as plt
```

```python
def transformation_1(p):
    x=p[0]
    y=p[1]
    x1=0.5*x
    y1=0.5*y
    return x1,  y1
```

```python
def transformation_2(p):
    x=p[0]
    y=p[1]
    x1=0.5*x+0.5
    y1=0.5*y+0.5
    return x1,  y1
```

```python
def transformation_3(p):
    x=p[0]
    y=p[1]
    x1=0.5*x+1
    y1=0.5*y
    return x1,  y1
```

```python
def get_index(probability):
    r=random.random()
    c_probability=0
    sum_probability=[]
    for p in probability:
        c_probability +=p
        sum_probability.append(c_probability)
    for item, sp in enumerate(sum_probability):
        if r<=sp:
            return item
    return len(probability)-1
```

```python
def transform(p):
    # list of transformation functions
    transformations=[transformation_1, transformation_2, transformation_3]
    probability=[0.3333, 0.3333, 0.3333]
    # pick a random transformation function and call it
    tindex=get_index(probability)
    t=transformations[tindex]
    x,y=t(p)
    return x,y
```

```python
def draw_Sierpinski(n):
    # we start with (0,0)
    x=[0]
    y=[0]
    
    x1, y1=0, 0
    for i in range(n):
        x1, y1=transform((x1,y1))
        x.append(x1)
        y.append(y1)
    return x,y
```

```python
def plot_Sierpinski(n):
    x,y=draw_Sierpinski(n)
    # plot the points
    plt.plot(x,y,'o')
    plt.title("Sierpinski gasket with {0} points".format(n))
    plt.show()
```

```python
plot_Sierpinski(1000)
```
